Outlier Detection and removal

```python
from scipy import stats
```

```python
# calculating z-score of oxy and deoxy columns
oxy_columns = [col for col in final_dataset1.columns if 'oxy' in col]
deoxy_columns = [col for col in final_dataset1.columns if 'deoxy' in col]
```

```python
#combing all oxy and deoxy columns
all_columns = oxy_columns + deoxy_columns
```

```python
#Calculating the z-score
z_score = stats.zscore(final_dataset1[all_columns])
```

```python
import numpy as np
abs_z_score = np.abs(z_score)
final_dataset1 = final_dataset1[(abs_z_score < 3).all(axis=1)]
```

```python
final_dataset1.shape
```

Feature Engineering

```python
# Create new columns for average oxy and deoxy levels across all channels
final_dataset1['avg_oxy'] = final_dataset1[oxy_columns].mean(axis=1)
final_dataset1['avg_deoxy'] = final_dataset1[deoxy_columns].mean(axis=1)
```

```python
# Optionally, calculate variance or other statistical measures
final_dataset['var_oxy'] = final_dataset[oxy_columns].var(axis=1)
final_dataset['var_deoxy'] = final_dataset[deoxy_columns].var(axis=1)
```
