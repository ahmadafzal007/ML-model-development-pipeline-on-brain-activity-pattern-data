Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
oxy_columns = [col for col in combined_brain_activity.columns if 'oxy' in col]
deoxy_columns = [col for col in combined_brain_activity.columns if 'deoxy' in col]
```

histogram for distribution

```python
plt.figure(figsize=(10, 10))
sns.histplot(combined_brain_activity[oxy_columns].values.flatten(), kde=True, color='blue', label='Oxy Levels')
sns.histplot(combined_brain_activity[deoxy_columns].values.flatten(), kde=True, color='red', label='Deoxy Levels')
plt.legend()
plt.title('Distribution of Oxy and Deoxy Levels across Channels')
plt.show()  
```

Pairplot or heatmap for Correlation Analysis

```python
correaltion_matrix = combined_brain_activity[oxy_columns + deoxy_columns].corr()    
plt.figure(figsize=(30, 25))
sns.heatmap(correaltion_matrix, cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Oxy and Deoxy Levels')
plt.show()
```

Boxplot for Outlier Detection

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=combined_brain_activity[oxy_columns].values)
plt.title('Boxplot of Oxy Levels across Channels')
plt.show()
```

Time Series Plot

```python
plt.figure(figsize=(12,6))
plt.plot(combined_brain_activity['row_id'], combined_brain_activity[oxy_columns[0]], label='Oxy Channel 1')
plt.xlabel('Row ID')
plt.ylabel
plt.title('Oxy Channel Over Time for Participant')
plt.legend()
plt.show()
```

# Participant data

```python
processed_participant_data = pd.read_csv('processed_participent_data.csv')
processed_participant_data.head()
```

summary of data

```python
print(processed_participant_data.info())
print(processed_participant_data.describe(include='all'))
```

Participants Group distribution

```python
sns.countplot(x='participant_group', data=processed_participant_data)
plt.title('Distribution of Participant Groups')
plt.xlabel('Participant Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Participant Age Distribution

```python
sns.histplot(processed_participant_data['participant_age'], kde=True, color='skyblue'), 
plt.title('Distribution of Participant Age')    
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Participant Sec Distribution

```python
# Gender distribution as a pie chart
sex_counts = processed_participant_data['participant_sex'].value_counts()
sex_counts.plot.pie(autopct='%1.1f%%', colors=['lightblue', 'pink'], labels=['Male', 'Female'], figsize=(6, 6))
plt.title('Participant Sex Distribution')
plt.ylabel('')  # Remove the default y-axis label
plt.show()
```

*** MOCA, RBANS, and CRIQ Scores ***

(a) Boxplot for Scores

```python
# Boxplot for MOCA, RBANS, and CRIQ scores
sns.boxplot(data=processed_participant_data[['participant_moca', 'participant_rbans', 'participant_criq']], palette='pastel')
plt.title('Boxplot of Cognitive Scores')
plt.xlabel('Scores')
plt.ylabel('Values')
plt.show()
```

(b) Histograms for Each Score

```python
# Histograms for cognitive scores
fig, axs = plt.subplots(1, 3, figsize=(30, 10))

sns.histplot(processed_participant_data['participant_moca'], kde=True, ax=axs[0], color='lightgreen')
axs[0].set_title('MOCA Score Distribution')
axs[0].set_xlabel('MOCA Score')

sns.histplot(processed_participant_data['participant_rbans'], kde=True, ax=axs[1], color='lightcoral')
axs[1].set_title('RBANS Score Distribution')
axs[1].set_xlabel('RBANS Score')

sns.histplot(processed_participant_data['participant_criq'], kde=True, ax=axs[2], color='lightblue')
axs[2].set_title('CRIQ Score Distribution')
axs[2].set_xlabel('CRIQ Score')

plt.tight_layout()
plt.show()
```

Relationship between age and scores

```python
# Scatterplot: Age vs MOCA
sns.scatterplot(x='participant_age', y='participant_moca', data=processed_participant_data, hue='participant_group', palette='coolwarm')
plt.title('Age vs MOCA Scores')
plt.xlabel('Age')
plt.ylabel('MOCA Score')
plt.legend(title='Group')
plt.show()

# Scatterplot: Age vs RBANS
sns.scatterplot(x='participant_age', y='participant_rbans', data=processed_participant_data, hue='participant_group', palette='viridis')
plt.title('Age vs RBANS Scores')
plt.xlabel('Age')
plt.ylabel('RBANS Score')
plt.legend(title='Group')
plt.show()
```

Distribution of Scores by Group

```python
# Boxplot: MOCA scores by group
sns.boxplot(x='participant_group', y='participant_moca', data=processed_participant_data, palette='pastel')
plt.title('MOCA Scores by Participant Group')
plt.xlabel('Participant Group')
plt.ylabel('MOCA Score')
plt.xticks(rotation=45)
plt.show()

# Boxplot: RBANS scores by group
sns.boxplot(x='participant_group', y='participant_rbans', data=processed_participant_data, palette='pastel')
plt.title('RBANS Scores by Participant Group')
plt.xlabel('Participant Group')
plt.ylabel('RBANS Score')
plt.xticks(rotation=45)
plt.show()

# Boxplot: CRIQ scores by group
sns.boxplot(x='participant_group', y='participant_criq', data=processed_participant_data, palette='pastel')
plt.title('CRIQ Scores by Participant Group')
plt.xlabel('Participant Group')
plt.ylabel('CRIQ Score')
plt.xticks(rotation=45)
plt.show()
```
